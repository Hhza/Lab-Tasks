```python
class Anime:

    def __init__(self, genre,country):
        self.genre = genre
        self.country = country
    def display(self):
        print(f"The anime is from {self.country} and its genres are {self.genre}",end="\n\n")

class Series(Anime):
    def __init__(self,name,no_of_ep,rating,genre,country):
        super().__init__(genre,country)
        self.name=name
        self.no_of_ep=no_of_ep
        self.rating=rating
    def display(self):
        Anime.display(self)
        print(f"The anime name is {self.name} its episodes are {self.no_of_ep} and its ratinf is {self.rating}",end=" " )
obj1=Anime(["shounen","adventure","kons"],"Japan")
obj1.display()        
obj2=Series("Jujutsu Kaisaen","24",8.9,["shounen","adventure","kons"],"Japan")
obj2.display()
```

```python
a=["shounen","adventure","konk","xanx","xander"]

def t2(a):
    a.sort()
    temp=[]
    temp2=[]
    count=0
    for i in a:
        if(i[0]=='X' or i[0]=='x'):
            temp.append(i)
            temp2.append(count)        
            count+=1
    for i in temp2:
        del a[i]
    return a,temp
a,b=t2(a)
print(a)
print(b)
```
